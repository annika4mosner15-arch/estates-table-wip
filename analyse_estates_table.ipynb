## Mapping

```python
"""
Dictionary mapping symbolic input patterns to corresponding encoded sequences.

Each key represents a specific symbol or combination of symbols, while the 
associated value is a list of one or more codes that encode or transform 
that input into a normalized or structured representation.
"""
mapping = {
    "x": ["a"],
    "+": ["b"],  
    "o": ["c"],
    "±": ["d"],

    "Fx": ["y", "a"],
    "F+": ["y", "b"],
    "F±": ["y", "d"],

    "++": ["b", "x"],
    "xx": ["a", "x"],
    "±+": ["d", "b"],

    "(x)": ["a", "z"],
    "(o)": ["c", "z"],
    "(±)": ["d", "z"],

    "+x": ["b", "a"],
    "ox": ["c", "a"]
}
```

```python
"""
This code takes a subset of columns (`rt_cols`) from the DataFrame `df`
and replaces each string value according to the predefined `mapping` dictionary.

Process:
- For each column in `rt_cols`, apply a transformation element-wise.
- For each cell value `x`:
    - If `x` is a string, look it up in `mapping`.
        - If a match is found, the corresponding list of codes is joined
          into a single comma-separated string (e.g., ["a", "b"] → "a,b").
        - If no match is found, an empty list is returned by default,
          resulting in an empty string.
    - If `x` is not a string (e.g., NaN or numeric), it is left unchanged.
Result:
- Returns a new DataFrame (`df_mapped`) with the same structure as the
  selected columns, but with symbolic values converted into their encoded
  string representations.
"""

df_mapped = df[rt_cols].apply(
    lambda col: col.map(
        lambda x: ",".join(mapping.get(x, [])) if isinstance(x, str) else x
    )
)
```

```python
"""
Map a single value `x` to its encoded representation using `mapping`.

- If `x` exists as a key in the `mapping` dictionary:
    - Retrieve the corresponding list of codes (e.g., ["a", "b"])
    - Join them into a comma-separated string (e.g., "a,b")
- If `x` is not found in `mapping`:
    - Return `x` unchanged
"""

lambda x: ",".join(mapping[x]) if x in mapping else x
```
